## Problema 2

```python
def reconstruir_imagen_color(ruta_azul, ruta_rojo, ruta_verde):
    imagen_azul = cv2.imread(ruta_azul, cv2.IMREAD_GRAYSCALE)
    imagen_rojo = cv2.imread(ruta_rojo, cv2.IMREAD_GRAYSCALE)
    imagen_verde = cv2.imread(ruta_verde, cv2.IMREAD_GRAYSCALE)

    imagen_color = np.zeros((imagen_azul.shape[0], imagen_azul.shape[1], 3), dtype=np.uint8)
    imagen_color[:, :, 0] = imagen_azul
    imagen_color[:, :, 1] = imagen_verde
    imagen_color[:, :, 2] = imagen_rojo

    return imagen_color
```

```python
ruta_imagen1_azul = 'imagen1/imagen1_salida_gray_azul.jpg'
ruta_imagen1_rojo = 'imagen1/imagen1_salida_gray_rojo.jpg'
ruta_imagen1_verde = 'imagen1/imagen1_salida_gray_verde.jpg'

ruta_imagen2_azul = 'imagen2/imagen2_salida_gray_azul.jpg'
ruta_imagen2_rojo = 'imagen2/imagen2_salida_gray_rojo.jpg'
ruta_imagen2_verde = 'imagen2/imagen2_salida_gray_verde.jpg'

ruta_perro_azul = 'perro/perro_salida_gray_azul.jpg'
ruta_perro_rojo = 'perro/perro_salida_gray_rojo.jpg'
ruta_perro_verde = 'perro/perro_salida_gray_verde.jpg'

imagen1_color = reconstruir_imagen_color(ruta_imagen1_azul, ruta_imagen1_rojo, ruta_imagen1_verde)
imagen2_color = reconstruir_imagen_color(ruta_imagen2_azul, ruta_imagen2_rojo, ruta_imagen2_verde)
perro_color = reconstruir_imagen_color(ruta_perro_azul, ruta_perro_rojo, ruta_perro_verde)

cv2.imwrite('imagen1_color.jpg', imagen1_color)
cv2.imwrite('imagen2_color.jpg', imagen2_color)
cv2.imwrite('perro_color.jpg', perro_color)
```

## Problema 4

```python
def mostrar_histogramas(imagen_color):
    azul, verde, rojo = cv2.split(imagen_color)
    gris = (azul.astype(np.uint16) + verde.astype(np.uint16) + rojo.astype(np.uint16)) // 3
    gris = gris.astype(np.uint8)

    canales = [azul, verde, rojo, gris]
    colores = ['b', 'g', 'r', 'k']
    nombres = ['Azul', 'Verde', 'Rojo', 'Gris']

    for canal, color, nombre in zip(canales, colores, nombres):
        hist, bins = np.histogram(canal.flatten(), 256, [0, 256])
        plt.plot(hist, color=color)
        plt.title(f'Histograma del canal {nombre}')
        plt.xlabel('Intensidad')
        plt.ylabel('Frecuencia')
        media = np.mean(canal)
        plt.axvline(media, color='k', linestyle='dashed', linewidth=1)
        plt.show()
```

```python
imagen_prueba = cv2.imread('img1.png')
mostrar_histogramas(imagen_prueba)
```

## Problema 5

```python
def escala_grises_ponderada(imagen_color):
    azul, verde, rojo = cv2.split(imagen_color)
    gris_ponderado = 0.7 * azul + 0.2 * verde + 0.1 * rojo
    gris_ponderado = gris_ponderado.astype(np.uint8)
    return gris_ponderado
```

```python
imagen_prueba = cv2.imread('img1.png')
imagen_gris_ponderada = escala_grises_ponderada(imagen_prueba)
cv2.imwrite('imagen_prueba_gris_ponderada.jpg', imagen_gris_ponderada)
```

## Problema 6

```python
def convertir_hsv(imagen_color):
    imagen_hsv = cv2.cvtColor(imagen_color, cv2.COLOR_BGR2HSV)
    return imagen_hsv

```

```python
imagen_prueba = cv2.imread('img1.png')
imagen_hsv = convertir_hsv(imagen_prueba)
cv2.imwrite('imagen_prueba_hsv.jpg', imagen_hsv)
```

El espacio de color HSV (Hue, Saturation, Value) es un modelo de color que describe los colores en trminos de su tono (Hue), saturacin (Saturation) y valor (Value). 
Este espacio de color es til porque separa la informacin de color (Hue) de la informacin de intensidad (Value), lo que facilita la manipulacin y anlisis de imgenes basadas en colores.

Hue representa el tipo de color y se mide en grados donde 0 es rojo, 120 es verde y 240 es azul.
Saturation representa la pureza del color y vara de 0% (gris) a 100% (color puro).
Value representa el brillo del color y vara de 0% (negro) a 100% (brillo mximo).
